# Importing the Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.metrics import accuracy_score, make_scorer
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

# Importing the dataset

```python
df = pd.read_csv(r"heart.csv")
```

```python
df
```

```python
scal=MinMaxScaler()
feat=['age', 	'sex', 	'cp', 'trestbps', 'chol', 	'fbs', 	'restecg', 	'thalach' ,	'exang', 	'oldpeak' ,	'slope', 	'ca', 'thal']
df[feat] = scal.fit_transform(df[feat])
df.head()
```

```python
x = df.drop(columns= ["target"], axis= 1)
y = df["target"]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 42)
```

```python
x_train
```

```python
y_train
```

```python
models = [LogisticRegression(), KNeighborsClassifier(n_neighbors= 4), RandomForestClassifier(n_estimators= 5),SVC(kernel= "linear")]
```

```python
for model in models:
    cross_val_results = cross_val_score(model, x, y, cv = 10)
    print(f'Cross-Validation Results for {model} is : {cross_val_results}')
    print(f'Mean Accuracy: {round(cross_val_results.mean()*100,2)}%')
    print("_________________________________________________________");
```

## Random forest has the highest accuracy

## HyperParamter tuning of Logistic Regression

```python
log = LogisticRegression()
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],  # Regularization parameter
    'penalty': ['l1', 'l2'],               # Penalty type
    'solver': ['liblinear', 'saga']       # Solver type
}
grid_search = GridSearchCV(estimator=log, param_grid=param_grid, cv=10, verbose=1)

grid_search.fit(x_train,y_train)

best_param = grid_search.best_estimator_
print("Best Hyperparameters:", grid_search.best_params_)

print("Accuracy:- ", best_param.score(x_test,y_test))
```

## Hyperparamter tuning of SVC

```python
svm = SVC()

param_grid = {
    'C': [0.1, 1, 10, 100],          # Regularization parameter
    'kernel': ['linear', 'rbf'],     # Kernel type
    'gamma': ['scale', 'auto']       # Kernel coefficient
}

grid_search1 = GridSearchCV(estimator=svm, param_grid= param_grid, cv = 10, verbose= 1)
grid_search1.fit(x_train, y_train)

best_param_svm = grid_search1.best_estimator_
print("best parameter:- ", grid_search1.best_params_)

print("Accuracy for the best parameter:- ", best_param_svm.score(x_test, y_test))
```

## Hyperparameter tuning of KNN

```python
knn = KNeighborsClassifier()

knn_grid={'n_neighbors': np.arange(1,30,1),
          'leaf_size': np.arange(1,50,1)}

gs_knn=GridSearchCV(KNeighborsClassifier(),param_grid=knn_grid,cv=5,verbose=True)

gs_knn.fit(x_train, y_train)

gs_knn.best_params_

print(f"Accuracy score:{gs_knn.score(x_test, y_test)*100}%")
```

## HyperParameter tuning of Random Forest

```python
rf = RandomForestClassifier()

param_grid = {
    'n_estimators': [100, 200, 300],    # Number of trees in the forest
    'max_depth': [None, 10, 20, 30],     # Maximum depth of the trees
    'min_samples_split': [2, 5, 10],     # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4]        # Minimum number of samples required to be at a leaf node
}

rf_grid = GridSearchCV(rf, param_grid= param_grid, cv = 5, verbose= True)

rf_grid.fit(x_train,y_train)

best_param_rf = rf_grid.best_estimator_
print("Best parameter:- " , rf_grid.best_params_)

print("accuracy of the model is:- ", best_param_rf.score(x_test, y_test))
```

```python
model_comp = pd.DataFrame({'Model': ['Logistic Regression','Random Forest',
                    'K-Nearest Neighbour','Support Vector Machine'], 'Accuracy': [best_param.score(x_test,y_test)*100,
                    best_param_rf.score(x_test,y_test)*100,gs_knn.score(x_test, y_test)*100,best_param_svm.score(x_test, y_test)*100]})
model_comp
```
